# Question 1

```python
# wind conditions based on current speed as Calm, Moderate or strong

def wind_condition(speed):
    if speed < 5:
        return "Calm"
    elif 5 <= speed <= 20:
        return "Moderate"
    else:
        return "Strong"
```

The function of wind_condition uses the argument of speed, to represent the wind speed in km/h. Determined by this value it returns a string based on whether its "Calm" less than 5km/h, "Moderate" between 5 and 20 km/h, and "Strong" above 20 km/h

There are multiple ways in which a function can allow for your code to be reusable.

Firstly functions allow for you to be able to call upon them later with no need to rewrite the code. Due to the nature of being separated from the rest of the code, it allows for the code to be more concise and clean, as well as being easier to trace any bugs or issues with the code. Because these functions can be considered as modular, you can reuse these functions in other programs. By changing out the parameters of function you can adapt the code for various other situations without needed to write an entirely new block of code.





# Question 2

```python
import matplotlib.pyplot as plt

# Sampled daily precipitation data for the first month
precipitation_data = [5.2, 7.3, 0, 0, 9.3, 0, 0, 0, 7.2, 0, 7, 0, 0, 0.6, 0, 0, 0, 0, 0, 0,
                      9.1, 0, 6.9, 0, 0, 0, 0, 0, 4.4, 0]

# Bar plot of daily precipitation levels
days = list(range(1, len(precipitation_data) + 1))
plt.figure(figsize=(10, 5))
plt.bar(days, precipitation_data, color='blue')
plt.title('Daily Precipitation Levels')
plt.xlabel('Day')
plt.ylabel('Precipitation (mm)')
plt.show()

# Classify each day based on precipitation level
no_rain = sum(1 for x in precipitation_data if x == 0)
light_rain = sum(1 for x in precipitation_data if 0 < x < 5)
heavy_rain = sum(1 for x in precipitation_data if x >= 5)

# Pie chart of precipitation categories
labels = ['No rain', 'Light rain', 'Heavy rain']
sizes = [no_rain, light_rain, heavy_rain]
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Proportion of Days by Rain Type')
plt.show()
```

# Pie Chart

The bar graph shows us a distribution of days it was not raining, lightly raining, and raining heavily as a percentage. We can see that no rain consists of 70% of the days within the first month. Light rain consists of 6.7% of the days within the first month. Heavy rain consists of 23.3% of the days in the first month.

This means that although for the most part it was not raining there was days in which it was raining and those days consisted of mostly heavy rain.

# Bar Graph

The bar graph shows us the daily amounts of rain per day in mm. Through this we can see how and when rainfall occurs to understand a trend. We can see that 6 of the 9 days that were raining occured within the first half of the month. By looking at the overall trend we can see that it looks similar to an inverted bell curve with high instances of rain on either side of the middle.

# Question 4

```python
import random

def monitor_temperatures():
    try:
        for hour in range(12):
            temperature = random.uniform(-5, 15)  # Simulating temperature reading
            print(f"Hour {hour + 1}: {temperature:.2f}°C")
            if temperature < 0:
                raise ValueError("Temperature fell below 0°C. Stopping further recordings.")
    except ValueError as e:
        print(e)

monitor_temperatures()
```

The way in which I handled the exception scenario is by having a try except block. Therefore, by raising a ValueError when the temperature is below 0 it exits the loop and stops monitoring hourly values.

```python
# Dictionary to store precipitation data for each city
precipitation_data = {
    "CityA": [12.3, 5.1, 0, 3.2],
    "CityB": [0, 7.8, 6.3, 0],
    "CityC": [3.5, 0, 8.1, 2.2],
    "CityD": [1.2, 4.7, 0, 9.0]
}

# Function to display precipitation data for a specific city
def display_city_precipitation(city):
    if city in precipitation_data:
        print(f"Precipitation data for {city}: {precipitation_data[city]}")
    else:
        print(f"City '{city}' not found in data.")

# Function to calculate the total precipitation for all cities
def calculate_total_precipitation():
    total_precipitation = 0
    for city, readings in precipitation_data.items():
        total_precipitation += sum(readings)
    return total_precipitation

# Display precipitation data for a specific city
city_to_display = "CityB"  # Change the city name to test with different cities
display_city_precipitation(city_to_display)

# Calculate and display the total precipitation for all cities
total_precipitation = calculate_total_precipitation()
print(f"\nTotal precipitation for all cities: {total_precipitation} mm")
```
